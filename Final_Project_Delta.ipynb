### Context

##### Apa Itu Customer Churn?
Customer churn merupakan penurunan pelanggan yang dapat dilihat/diukur dari persentase jumlah pelanggan yang berhenti membeli produk atau layanan bisnis Anda dalam jangka waktu tertentu. Apabila bisnis Anda mendapat customer churn rate yang tinggi, maka artinya ada banyak customer Anda yang berhenti berlangganan dengan bisnis Anda. Maka tidak heran jika kondisi yang satu ini sangat ingin dihindari semua bisnis.

Jika bisnis Anda memiliki churn rate yang tinggi. Hal tersebut bisa disebabkan oleh berbagai macam, seperti ketidakpuasan pelanggan terhadap produk atau layanan bisnis Anda. Penyebab lainnya juga bisa datang dari kegiatan marketing bisnis Anda yang kurang sempurna, contohnya target dan metode pemasaran product yang kurang tepat.

Jenis-Jenis Customer Churn: 
- Voluntary churn adalah Pelanggan secara sadar memutuskan untuk berhenti menggunakan layanan atau produk, biasanya karena alasan seperti harga, kualitas, atau layanan yang tidak memuaskan. Contoh : Jika ada seseorang berlangganan perbulan untuk Netflix, tetapi memutuskan untuk memutus langganan dikarenakan jarang menggunakan layanan Netflix.
- Involuntary churn adalah pemutusan berhenti menggunakan layanan yang terjadi karena hal-hal di luar keinginan pelanggan, misalnya kartu kredit gagal ditagih, atau akun terblokir karena alasan teknis. Contoh : Jika ada seseorang berlangganan perbulan untuk Netflix, tetapi berhenti langganan dikarenakan proses pembayaran gagal dan tidak ada cadangan pembayaran yang lainnya.

### Metric Evaluation
|           | Prediksi 1 | Prediksi 0 |
|-----------|--------|--------|
| **Aktual 1** | **True Positive (TP)**<br>Model memprediksi pelanggan churn dan memang benar begitu. | **False Negative (FN)**<br>Model memprediksi pelanggan tidak churn padahal sebenarnya churn. |
| **Aktual 0** | **False Positive (FP)**<br>Model memprediksi pelanggan churn padahal sebenarnya tidak churn.  | **True Negative (TN)**<br>Model memprediksi pelanggan tidak churn dan memang benar begitu. |

Konsekuensi dari dua jenis error:
- False Negative (FN): biaya penggantian pendapatan langsung dari pelanggan yang pergi ditambah biaya penggantian pelanggan atau CAC (Customer Acquisition Cost) yang tinggi. Pada tahun 2024 rata-rata CAC untuk Kategori Electronics : 76 USD /pelanggan, Kategori Fashion / Apparel : 66 USD/pelanggan, Kategori Food & Beverage : 53 USD/pelanggan.

- False Positive (FP): biaya penggantian dari penggunaan biaya retensi atau CRC (Customer Retention Cost) yang tidak tepat sasaran. Rata-rata biaya CRC yang dapat dikeluarkan untuk kategori Electronics adalah sekitar 15.2 USD/pelanggan , kategori Fashion / Apparel adalah sekitar 13.2 USD/pelanggan , dan kategori Food & Beverage adalah sekitar 10.6 USD/pelanggan .

Dari konsekuensi kedua error diatas, didapati bahwa FN lebih merugikan perusahaan dibandingkan FP. Oleh karena itu, metrik yang digunakan untuk mengevaluasi model adalah Recall kelas positif (pelanggan yang churn). Fokus terhadap metrik tersebut akan mengurangi FN sehingga pelanggan yang sebenarnya churn dapat teridentifikasi. Berikut Persamaan dari Recall:

$$
\text{Recall} (+) = \frac{TP}{TP+FN}
$$

### Pengecekan Outlier

```python
dataOutliers = []

for col in numerical_columns:
    if df[col].nunique() < 15:
        continue
        
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    batas_bawah = Q1 - 1.5 * IQR
    batas_atas  = Q3 + 1.5 * IQR
    outliers = df[(df[col] < batas_bawah) | (df[col] > batas_atas)]
    persen_outliers =  round((len(outliers)/len(df)) *100,4)

    batas_ekstrem_bawah = Q1 - 3 * IQR
    batas_ekstrem_atas = Q3 + 3 * IQR
    outliers_ekstrem = df[(df[col] < batas_ekstrem_bawah) | (df[col] > batas_ekstrem_atas)]
    persen_outliers_ekstrem =  round((len(outliers_ekstrem)/len(df)) *100,4)
    
    dataOutliers.append([
        col,
        len(outliers),
        persen_outliers,
        len(outliers_ekstrem),
        persen_outliers_ekstrem
    ])

pd.DataFrame(dataOutliers, columns=[
    'Data Features',
    'Outliers',
    'Outliers Percentage',
    'Extreme Outliers',
    'Extreme Outliers Percentage'
])
```

- Dilakukan pengecekan outlier untuk data numerik dengan nilai unik lebih dari atau sama dengan 15.
- Seluruh kolom memiliki outlier dengan persentase yang berbeda.
  - Sedikit: Tenure, WarehouseToHome, NumberOfAddress, dan OrderAmountHikeFromlastYear.
  - Sedang: CashbackAmount
  - Tinggi: CouponUsed, OrderCount
- Diputuskan untuk tidak melakukan penanganan outlier sehingga digunakan RobustScaler untuk scaling pada model KNN dan Logistic Regression yang sensitif terhadap skala.
- Model yang berbasis pohon tidak memerlukan scaling.
- Penanganan outlier akan dilakukan apabila seluruh model memiliki performa yang buruk.

#### Pengecekan Hubungan Categorical Features dengan Churn

```python
categorical_columns = df.select_dtypes(include=['O']).columns.tolist()+['Complain', 'CityTier', 'SatisfactionScore']

for col in categorical_columns:
    contingency_table = pd.crosstab(df[col], df['Churn'])
    
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    
    print(f'Fitur: {col}')
    print(f'P-value: {p:.4f}')
    
    if p < 0.05:
        print(f'Hasil: Ada hubungan signifikan antara {col} dan Churn\n')
    else:
        print(f'Hasil: Tidak ada hubungan signifikan antara {col} dan Churn\n')
```

- Dilakukan chi square test untuk mengecek hubungan categorical features dengan Churn.
- Seluruh categorical features memiliki p-value kurang dari 0.05 artinya ada hubungan signifikan antara setiap feature dan Churn.

```python
def ChurnPercent(feature):
    stats = (
        df.groupby(feature)['Churn']
        .agg(['mean', 'count'])
        .rename(columns={'mean': 'Churn %', 'count': 'Count'})
    )
    stats['Churn %'] = (stats['Churn %'] * 100).round(2)
    return stats
```

Regular function ChurnPercent digunakan untuk melihat hubungan categorical features dengan Churn secara lebih detail.

## Data Preprocessing

```python
cekUnik(df.columns)
```

Dari informasi nilai unik dan tipe data dari setiap kolom didapati:
- Numerik Biner: Complain. Fitur kategorikal yang nilainya sudah numerik.
- Ordinal: CityTier.
- Obyek Biner: PreferredLoginDevice dan Gender. Fitur kategorikal dengan nilai unik 2 yang tipe datanya obyek.
- Nominal: PreferredPaymentMode, PreferedOrderCat, dan MaritalStatus. Fitur kategorikal dengan nilai unik lebih dari 2.
- Numerik: Fitur dengan nilai numerik yang bukan Ordinal dan Numerik Biner.
- SatisfactionScore tidak digunakan sebagai fitur seperti yang dijelaskan pada tahapan EDA.

```python
num_binary_ordinal_cols = ['CityTier', 'Complain']

unscaled_cols = num_binary_ordinal_cols + ['Churn', 'SatisfactionScore']
num_cols = [col for col in numerical_columns if col not in unscaled_cols]

obj_binary_cols = ['PreferredLoginDevice', 'Gender']

object_cols = df.select_dtypes(include=['O']).columns.tolist()
obj_nominal_cols = [col for col in object_cols if col not in obj_binary_cols]
```

- Pengelompokan fitur-fitur berdasarkan perlakuan ketika data conversion.
- Fitur numerik biner dan ordinal dijadikan satu karena mendapat perlakuan yang sama.

```python
binary_encoder = OrdinalEncoder(categories=[['Mobile', 'Computer'],
                                            ['Female', 'Male']])

transformer_scaled = ColumnTransformer([
    ('obj_nominal', OneHotEncoder(), obj_nominal_cols),
    ('obj_binary', binary_encoder, obj_binary_cols),
    ('num_binary_ordinal', 'passthrough', num_binary_ordinal_cols),
    ('num', RobustScaler(), num_cols)
])

transformer_no_scale = ColumnTransformer([
    ('obj_nominal', OneHotEncoder(), obj_nominal_cols),
    ('obj_binary', binary_encoder, obj_binary_cols),
    ('num_binary_ordinal', 'passthrough', num_binary_ordinal_cols),
    ('num', 'passthrough', num_cols)
])
```

- OneHotEncoder untuk fitur nominal.
- binary_encoder untuk fitur obyek biner.
- Pada binary_encoder, urutan pengisian categories di OrdinalEncoder mengikuti urutan di obj_binary_cols. Kemudian, nilai yang lebih dulu akan menjadi 0 sehingga nanti hasilnya:
  - PreferredLoginDevice: 0 = "Mobile" dan 1 = "Computer".
  - Gender: 0 = "Female" dan 1 = "Female"
- Tidak dilakukan perubahan pada fitur numerik biner dan ordinal.
- Terdapat dua transformer yaitu transformer_scaled dan transformer_no_scale. Model KNN dan Logistic Regression membutuhkan scaling karena sensitif terhadap skala sehingga dibuat dua transformer.
- transformer_scaled menjadi transformer yang menyertakan scaling terhadap fitur numerik dengan RobustScaler.
- RobustScaler yang memanfaatkan nilai median dan IQR dalam scaling dipilih karena terdapat outlier pada fitur numerik.

#### Evaluasi Model XGBoost & Decision Tree
Dilakukan evaluasi terhadap Model XGBoost & Decision Tree untuk mengetahui apakah terdapat indikasi overfitting ataupun underfitting.

```python
def Eva_metrics_base_class(model, X_train, X_test, y_train, y_test, Nama, fit=False):
    if fit == False:
        Model = model.fit(X_train, y_train)
    else:
        Model = model
    y_pred_train = Model.predict(X_train)
    Acc_train = accuracy_score(y_train, y_pred_train)
    Rec_train = recall_score(y_train, y_pred_train)
    Prec_train = precision_score(y_train, y_pred_train)
    F1_train = f1_score(y_train, y_pred_train)
    
    y_pred_test = Model.predict(X_test)
    Acc_test = accuracy_score(y_test, y_pred_test)
    Rec_test = recall_score(y_test, y_pred_test)
    Prec_test = precision_score(y_test, y_pred_test)
    F1_test = f1_score(y_test, y_pred_test)
    
    data = {
        Nama + ' Training' : [Acc_train, Rec_train, Prec_train, F1_train],
        Nama + ' Testing' : [Acc_test, Rec_test, Prec_test, F1_test]
    }
    df = pd.DataFrame(data, index=['Accuracy', 'Recall', 'Precision', 'F1'])
    
    cr_tr = classification_report(y_train, y_pred_train)
    cm_tr = confusion_matrix(y_train, y_pred_train, labels=[1, 0])
    df_tr = pd. DataFrame(data=cm_tr, columns=['Pred 1', 'Pred 0'], index=['Akt 1', 'Akt 0' ])
    
    cr_ts = classification_report(y_test, y_pred_test)
    cm_ts = confusion_matrix(y_test, y_pred_test, labels=[1, 0])
    df_ts = pd. DataFrame(data=cm_ts, columns=['Pred 1', 'Pred 0'], index=[ 'Akt 1', 'Akt 0'])
    
    return df, cr_tr, df_tr, cr_ts, df_ts
```

Regular function Eva_metrics_base_class untuk menampilkan metrik untuk evaluasi, classification report, dan confusion matrix.

```python
xgb = XGBClassifier(random_state=42)

pipeline_xgb = Pipeline([
    ('preprocess', transformer_no_scale),
    ('model', xgb)
])

df_eval_xgb, cr_train_xgb, cm_train_xgb, cr_test_xgb, cm_test_xgb = Eva_metrics_base_class(
    pipeline_xgb,
    X_train, X_test,
    y_train, y_test,
    'XGBoost Base'
)
```

```python
dt = DecisionTreeClassifier(random_state=42)

pipeline_dt = Pipeline([
    ('preprocess', transformer_no_scale),
    ('model', dt)
])

df_eval_dt, cr_train_dt, cm_train_dt, cr_test_dt, cm_test_dt = Eva_metrics_base_class(
    pipeline_dt,
    X_train, X_test,
    y_train, y_test,
    'Decision Tree Base'
)
```

### Hyperparameter Tuning

```python
skfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

pipe_xgb = Pipeline([
    ('preprocess', transformer_no_scale),
    ('model', xgb)
])

params_xgb = {
    'model__scale_pos_weight': [1, 5, 10],
    'model__n_estimators': [100, 200, 300],
    'model__max_depth': [3, 5, 7],
    'model__learning_rate': [0.01, 0.05, 0.1],
    'model__subsample': [0.8, 1.0],
    'model__colsample_bytree': [0.8, 1.0],
    'model__reg_alpha': [0, 0.1, 1, 5],
    'model__reg_lambda': [1, 5, 10],
    'model__gamma': [0, 0.5, 1]
}

rs_xgb = RandomizedSearchCV(
    estimator=pipe_xgb,
    param_distributions=params_xgb, 
    n_iter=150,  
    cv=skfold,
    scoring='recall', 
    n_jobs=-1,
    verbose=2,
    random_state=42
)

rs_xgb.fit(X_train, y_train)

print('Best params:', rs_xgb.best_params_)
print('Best CV recall:', rs_xgb.best_score_)
```

Hal ini dilakukan karena parameter yang cukup banyak sehingga RandomizedSearchCV akan mempercepat proses hyperparameter tuning.
- Parameter yang digunakan untuk hyperparameter tuning:
  - scale_pos_weight: Menyeimbangkan bobot antara kelas positif dan negatif, dapat digunakan untuk data imbalance.
  - n_estimators: Menentukan jumlah boosting rounds yaitu jumlah pohon yang akan dibangun secara berurutan dalam proses boosting. Semakin besar nilainya maka dapat meningkatkan performa tapi juga risiko overfitting.
  - max_depth: Kedalaman maksimum dari setiap pohon. Semakin besar maka model akan semakin kompleks dan rawan overfit.
  - learning_rate: Ukuran langkah (step size) setiap boosting iteration yang akan menentukan seberapa cepat model belajar. Nilai yang kecil akan membuat belajar lambat namun lebih stabil.
  - subsample: Proporsi sampel data yang digunakan untuk membangun tiap tree.
  - colsample_bytree: Proporsi fitur yang digunakan untuk tiap tree.
  - reg_alpha: Regularisasi L1 yaitu menambah penalti terhadap bobot absolut sehingga dapat membantu mengurangi kompleksitas model. Semakin besar maka efek regularisasi semakin kuat.
  - reg_lambda: Regularisasi L2 yaitu memberikan penalti terhadap kuadrat bobot sehingga dapat membantu menstabilkan model & mencegah overfitting. Semakin besar maka efek regularisasi semakin kuat.
  - gamma: Minimum loss reduction yang dibutuhkan agar split dilakukan. Semakin besar model lebih konservatif karena lebih sedikit split sehingga mencegah overfitting.

#### Mengurangi Kerugian Akibat Churn
Pada tahapan ini kita dapat melihat bagaimana penggunaan model prediksi churn dapat mengurangi kerugian dari pelanggan churn. Kerugian yang ada berasal dari biaya penggantian pelanggan yang churn atau CAC (Customer Acquisition Cost) dan penggunaan biaya retensi atau CRC (Customer Retention Cost) yang tidak tepat sasaran. 

Kemudian, seperti yang dijelaskan di bagian Context bahwa CAC 5-25 kali lebih mahal dari CRC. Dipilih rasio terkecil yaitu 5 untuk menekankan dampak kerugian dari False Negative yaitu CAC jauh lebih mahal dibandingkan CRC. Dengan demikian, meskipun digunakan rasio minimum, hasil perhitungan tetap mencerminkan bahwa kerugian dari False Negative lebih signifikan. Dikarenakan kami hanya mendapatkan nilai rerata dari CAC per pelanggan maka rasio ini akan digunakan untuk mendapakan CRC per pelanggan. Nilai dari CRC per pelanggan memenuhi persamaan berikut:
$$
\text{CRC}_{\text{per Pelanggan}} = \frac{\text{CAC}_{\text{per Pelanggan}}}{5}
$$

Terdapat dua skenario ketika perusahaan tidak memiliki model prediksi:
1. Perusahaan akan memberikan usaha retensi pada seluruh pelanggan sehingga kerugiannya terdapat pada CRC yang salah sasaran ke pelanggan yang tidak berpotensi churn. Kerugian pada skenario ini memenuhi persamaan berikut:
$$
\text{Kerugian Tanpa Model Skenario 1} = \text{Not Churn}  \times \text{CRC}_{\text{per Pelanggan}}
$$

3. Perusahaan akan memberikan usaha retensi sebanyak jumlah pelanggan churn yang didapat dari data sekarang. Usaha tersebut akan diberikan secara random pada pelanggan perusahaan sehingga terdapat pelanggan berpotensi churn yang tidak mendapat usaha retensi dan pelanggan yang tidak berpotensi namun mendapat usaha retensi. Kerugian pada skenario ini memenuhi persamaan berikut:
$$
\text{CRC}_{\text{Salah Sasaran}} = \frac{\text{Churn}}{\text{Total Pelanggan}} \times \text{Not Churn} \times \text{CRC}_{\text{per Pelanggan}}
$$

$$
\text{CAC}_{\text{Penggantian}} = \frac{\text{Not Churn}}{\text{Total Pelanggan}} \times \text{Churn} \times \text{CAC}_{\text{per Pelanggan}}
$$

$$
\text{Kerugian Tanpa Model Skenario 2} = \text{CRC}_{\text{Salah Sasaran}} + \text{CAC}_{\text{Penggantian}}
$$

Apabila menggunakan model kerugian yang didapat berasal dari false negative sehingga perlu CAC untuk mengganti pelanggan yang pergi dan false positive yang menyebabkan penggunaan CRC yang salah sasaran. Kerugian ketika menggunakan model memenuhi persamaan berikut:
$$
\text{CRC}_{\text{FP}} = \text{FP} \times \text{CRC}_{\text{per Pelanggan}}
$$

$$
\text{CAC}_{\text{FN}} = \text{FN} \times \text{CAC}_{\text{per Pelanggan}}
$$

$$
\text{Kerugian dengan Model} = \text{CRC}_{\text{FP}} + \text{CAC}_{\text{FN}}
$$



```python
TN = cm_test_xgb_tuned.loc['Akt 0', 'Pred 0']
TP = cm_test_xgb_tuned.loc['Akt 1', 'Pred 1']
FN = cm_test_xgb_tuned.loc['Akt 1', 'Pred 0']
FP = cm_test_xgb_tuned.loc['Akt 0', 'Pred 1']

Churn = TP + FN
Not_Churn = TN + FP

def errorCost(Kategori, CAC):
    CRC = CAC / 5

    baseline_1_loss = Not_Churn * CRC
    
    baseline_cac_fn_cost = Not_Churn / (Churn + Not_Churn) * Churn * CAC
    baseline_crc_fp_cost = Churn / (Churn + Not_Churn) * Not_Churn * CRC
    baseline_2_loss = baseline_cac_fn_cost + baseline_crc_fp_cost
    
    model_cac_fn_cost = FN * CAC
    model_crc_fp_cost = FP * CRC
    model_loss = model_cac_fn_cost + model_crc_fp_cost

    loss_reduction_pct_1 = (baseline_1_loss - model_loss) / baseline_1_loss * 100
    loss_reduction_pct_2 = (baseline_2_loss - model_loss) / baseline_2_loss * 100

    print('Kategori:', Kategori)
    print(f'Total kerugian tanpa model pada skenario 1 adalah {baseline_1_loss:.3f} USD')
    print(f'Total kerugian tanpa model pada skenario 2 adalah {baseline_2_loss:.3f} USD')
    print(f'Total kerugian dengan model adalah {model_loss:.3f} USD\n')

    print(f'Skenario 1: Penggunaan model dapat mengurangi kerugian sebesar {loss_reduction_pct_1:.3f}%')
    print(f'Skenario 2: Penggunaan model dapat mengurangi kerugian sebesar {loss_reduction_pct_2:.3f}%')
```

#### Menurunkan Persentase Churn
Selain mengurangi kerugian, model ini dapat membantu perusahaan dalam mengurangi pelanggan churn. Dalam hal ini, terdapat dua skenario:
1. Optimis, seluruh usaha retensi ke pelanggan yang diprediksi churn berhasil sehingga mereka tidak churn. Persentase dari skenario optimis memenuhi persamaan berikut:
$$
\text{Churn \%}_{\text{Optimis}} = \frac{\text{FN}}{\text{Total Pelanggan}} \times 100
$$
3. Pesimis, hanya 50% usaha retensi ke pelanggan yang diprediksi churn berhasil sehingga hanya setengah dari mereka yang tidak churn. Persentase dari skenario pesimis memenuhi persamaan berikut:
$$
\text{Churn \%}_{\text{Pesimis}} = \frac{\text{FN} + \text{TP} \times 50\%}{\text{Total Pelanggan}} \times 100
$$

```python
churn_rate_optimis = FN / (Churn + Not_Churn) * 100
churn_rate_pesimis = (FN + TP * 50 / 100) / (Churn + Not_Churn) * 100

print('Churn rate awal = 16.8%')
print(f'Churn rate dengan model (Optimis) = {churn_rate_optimis:.3f}%')
print(f'Churn rate dengan model (Pesimis) = {churn_rate_pesimis:.3f}%')
```

- Menurut analisis pelanggan yang belum lama menggunakan layanan M-Mart juga memiliki kecenderungan untuk mudah churn. Ini mungkin dikarenakan kurangnya promo-promo ataupun benefit-benefit yang diberikan oleh M-Mart untuk penglanggan yang hanya memiliki pengalaman beberapa kali belanja di M-Mart. Maka dari itu saran strategi pemasaran yang kami berikan adalah adanya :
  1. Melakukan promo-promo cashback yang menarik untuk pelanggan-pelanggan yang masih berbelanja dalam jumlah yang sedikit sehingga menaikkan keinginan pelanggan dalam berbelanja pada awal-awal pelanggan menerima pelayanan berbelanja di e-commerce M-Mart.
  2. Konten-konten yang menarik selain memberikan promo voucher dan gratis ongkir.<br>
     Konten-konten yang menarik perhatian bisa di dapatkan dari :
     - Aktif melakukan pemasaran melalui Sosial Media. Sosial media dapat menarik keinginan beli pelanggan karena visual product yang di tampilkan
       dengan caption yang menarik perhatian. Terutama untuk barang-barang kategori Fashion.
     - Live Streaming untuk melakukan promosi penjualan. Cara ini sangat berguna untuk barang-barang dengan kategori electronik karena dapat sekalian
       mempraktekkan fungsi dan kegunaan dari barang elektronik yang di tawarkan beserta menampilkan keunggulan-keunggulan yang di tawarkan oleh barang

     - Menampilkan Ulasan dan Penilaian Pelanggan merupakan strategi yang dapat membangun kepercayaan calon pembeli. Ulasan positif

  3. Menaikkan pelayanan pelanggan (Customer Service)<br>

     Menguatkan brand awareness adalah langkah krusial dalam membangun fondasi yang kuat untuk bisnis E-Commerce. Kegunaan dari konten-konten yang
     menarik diatas adalah untuk memperkuat brand awareness yang sudah ada saat ini, selain itu brand awareness bisa didapat juga dengan menjalin
